```python
# --- クラス定義　---
class BankAccount:
    def __init__(self, initial_balance=0):
        # 残高は外部から直接変更できないようにカプセル化
        self.__balance = initial_balance

    def deposit(self, amount):
        """預け入れ処理"""
        if amount <= 0:
            print("エラー: 正しい金額を入力してください。")
            return
        self.__balance += amount
        print(f"{amount}円を預け入れました。現在の残高: {self.__balance}円")

    def withdraw(self, amount):
        """引き出し処理"""
        if amount <= 0 or amount > self.__balance:
            print("エラー: 残高不足または無効な金額です。")
            return
        self.__balance -= amount
        print(f"{amount}円を引き出しました。現在の残高: {self.__balance}円")

    def get_balance(self):
        """残高を返す"""
        return self.__balance


# --- 対話型処理 ---
account = BankAccount(1000)  # 初期残高1000円
print("銀行口座管理プログラム（初期残高: 1000円）")

while True:
    print("\n--- メニュー ---")
    print("1: 預け入れ")
    print("2: 引き出し")
    print("3: 残高照会")
    print("4: 終了")

    choice = input("操作を選んでください（1-4）: ")

    if choice == "1":
        try:
            amount = int(input("預け入れ金額を入力してください: "))
            account.deposit(amount)
        except ValueError:
            print("エラー: 数字を入力してください。")
    elif choice == "2":
        try:
            amount = int(input("引き出し金額を入力してください: "))
            account.withdraw(amount)
        except ValueError:
            print("エラー: 数字を入力してください。")
    elif choice == "3":
        print(f"現在の残高: {account.get_balance()}円")
    elif choice == "4":
        print("プログラムを終了します。")
        break
    else:
        print("エラー: 1〜4の数字を入力してください。")
```
